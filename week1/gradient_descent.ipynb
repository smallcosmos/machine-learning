# 使用plt绘制散点图

```python
import matplotlib.pyplot as plt
import numpy as np

# 绘制散点图
plt.scatter(np.array([1, 2]), np.array([300, 500]), marker='x', c='r')
plt.title('Housing Prices')
plt.xlabel('Size (1000 sqft)')
plt.ylabel('Price (in 1000s of dollars)')

plt.show()
```

# 使用plt绘制线图 & 散点图

```python
def compute_model_output(x, w, b):
  m = x.shape[0]
  f_wb = np.zeros(m)
  for i in range(m):
    f_wb[i] = w * x[i] + b

  return f_wb

x_train = np.array([1,2])
temp_f_wb = compute_model_output(x_train, 100, 100)

# 绘制线图
plt.plot(np.array([1, 2]), temp_f_wb, c='b', label='Actual Values')

# 绘制散点图
plt.scatter(np.array([1, 2]), np.array([300, 500]), marker='x', c='r')
plt.title('House Prices')
plt.xlabel('House Size')
plt.ylabel('Price')
```

# 定义方差函数

```python
from lab_utils_uni import plt_intuition

x_train = np.array([1,2])
y_train = np.array([300, 500])
def compute_cost(x, y, w, b):
  m = x.shape[0]
  cost_sum = 0
  for i in range(m):
    f_wb = w * x[i] + b
    cost_sum += (y[i] - f_wb) ** 2
  total_cost = 1 / (2 * m) * cost_sum

  return total_cost

test_cost = compute_cost(x_train, y_train, 100, 100)
print(test_cost)

plt_intuition(x_train, y_train)
```
